```python
df = pd.read_csv('final_vg.csv')
```

```python
df['Genre'].value_counts()
```

```python
cat_counts = df['Genre'].value_counts()
```

```python
x = cat_counts.index
y = cat_counts

plt.bar(x, y);
```

```python
plt.figure(figsize=(12, 10))
x = cat_counts.index
y = cat_counts

plt.bar(x, y);
plt.xticks(rotation = 90, fontsize = 12)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
x = cat_counts.index
y = cat_counts

plt.barh(x, y);
plt.xticks(rotation = 90, fontsize = 12)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
x = cat_counts.index
y = cat_counts

plt.bar(x, y);
plt.xticks(rotation = 45, fontsize = 12)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
x = cat_counts.index
y = cat_counts

plt.bar(x, y, width=.7, )

plt.xticks(rotation = 45, fontsize = 12)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
x = cat_counts.index
y = cat_counts

plt.bar(x, y, 
        width=.7,
        color = 'green')

plt.xlabel('Genre', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title("Count of Genre", fontsize=20)

plt.xticks(rotation = 45, fontsize = 12)
plt.show()
```

```python
plt.figure(figsize=(12, 6))

sns.countplot(data = df, 
              x = 'Genre')

plt.xticks(rotation = 45, fontsize = 12)
plt.title("Count of Genre", fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(12, 6))

sns.countplot(data = df, 
              x = 'Genre',
              order = df['Genre'].value_counts().index)

plt.xticks(rotation = 45, fontsize = 12)
plt.title("Count of Genre", fontsize=20)
plt.show()
```
